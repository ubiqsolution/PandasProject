```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')

import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams.update({'font.size': 14})

%matplotlib inline
```

```python
df = pd.read_csv('seed_sales_data.csv', encoding='cp949')
df.shape
```

```python
#plt.figure(figsize=(20,5))
plt.figure(figsize=(300,150))
plt.title('날짜별 주문건수')
plt.xticks(rotation=60, ha='right')
sns.countplot(data=df, x="날짜")
```

```python
figure,(ax1) = plt.subplots(1)
figure.set_size_inches(7,225)
#날짜별
sns.countplot(data=df, y="날짜", ax=ax1)
```

```python
figure,(ax1) = plt.subplots(1)
figure.set_size_inches(8,250)
#상품별
sns.countplot(data=df, y="상품명", ax=ax1)
```
