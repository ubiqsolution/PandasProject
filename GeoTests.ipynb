### * 위경도 변환 테스트!

```python
import pandas as pd
import numpy as np
import geopy
from geopy.geocoders import Nominatim
```

```python
# csv 파일 불러오기
csv = pd.read_csv('mgc.csv')
csv.head()
```

```python
# 데이터프레임 주소값 추출
address = csv['주소']
address.head()
```

```python
# 주소 데이터 깔끔하게 다듬기
for i in range(len(address)):
    a = address[i].split(' ')
    address[i] = " ".join(a[0:4])
print(address)
```

```python
geo_local = Nominatim(user_agent='South Korea')
```

```python
# 위도, 경도 반환하는 함수
def eval_results(x):
    try:
        return (x.latitude, x.longitude)
    except:
        return (None, None)

geo_df = address.apply(geo_local.geocode, timeout=10).apply(lambda x: eval_results(x))
```

```python
latitude = []
longitude = []

for i in geo_df:
    latitude.append(geo_df(i)[0])
    longitude.append(geo_df(i)[1])
```

```python
# Dataframe 만들기
address_geo = pd.DataFrame({'지역': csv['지역'],'매장명': csv['매장명'],'주소' : address, '위경도' : geo_df})
```

```python
# Dataframe 저장 
address_geo.to_csv('mgc_geo.csv', index = False)
```
