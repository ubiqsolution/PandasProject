```python
figure,(ax1, ax2) = plt.subplots(1, 2)
figure.set_size_inches(20,10)
# 상품별 주문수 1
sns.countplot(data=df, y="상품명", ax=ax1)
# 상품별 주문수 2
sns.barplot(data=df, x="상품명", y="주문수", ax=ax2)
```

```python
plt.figure(figsize=(20,5))
plt.title('날짜별 주문건수')
plt.xticks(rotation=60, ha='right')
sns.countplot(data=df, x="날짜")
```

```python
figure,(ax1) = plt.subplots(1)
figure.set_size_inches(10,5)
#업체별 주문수
sns.countplot(data=df, y="업체명", ax=ax1)
```
