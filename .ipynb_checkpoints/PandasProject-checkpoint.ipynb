## ★캠핑용품 블랙키위 키워드 데이터 분석!

```python
# 차후 데이터셋 다운로드 변경 요망.
# df = pd.read_csv('2021_12_19_camping.csv', encoding='cp949') encoding 에러!
df = pd.read_csv('2021_12_19_camping.csv')
df.shape
```

```python
df['월간 검색량 (전체)'].value_counts()
```

```python
df['키워드'].value_counts()
```

```python
figure,(ax1, ax2) = plt.subplots(1, 2)
figure.set_size_inches(30,20)
# 키워드
sns.countplot(data=df, y="키워드", ax=ax1)
# 키워드별 월간 검색량 (전체)
sns.barplot(data=df, x="키워드", y="월간 검색량 (전체)", ax=ax2)
```

```python
plt.figure(figsize=(30,15))
plt.title('주요 키워드별 월간 검색량(전체)')
plt.xticks(rotation=60, ha='right')
sns.countplot(data=df, x="월간 검색량 (전체)")
```

```python
figure,(ax1) = plt.subplots(1)
figure.set_size_inches(10,5)
# 키워드
sns.countplot(data=df, y="키워드", ax=ax1)
```
